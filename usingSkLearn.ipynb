```python
# Importing the numpy library
import numpy as np
```

```python
# Setting the number of data points (m) to 100
m = 100
# Generating random data for X using a uniform distribution and scaling it
X = 6 * np.random.rand(m, 1) - 3
# Generating the corresponding y values with a quadratic relationship and adding some noise
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)
```

```python
# Importing the pyplot module from matplotlib and aliasing it as plt
from matplotlib import pyplot as plt 
plt.scatter(X,y, color='blue')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# Transforming the original feature matrix X into polynomial features
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree = 2, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

```python
# Fitting the linear regression model on the polynomial features
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
# Plotting the predicted values on top of the original data points
X_poly = poly_features.transform(sorted(X))
y_pred = lin_reg.predict(X_poly)
plt.plot(sorted(X),y_pred, color='red')
plt.scatter(X,y)
```
